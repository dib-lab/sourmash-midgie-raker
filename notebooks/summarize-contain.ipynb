```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('../outputs.ath/contain/summary.csv')
```

```python
df
```

```python
plt.plot(df.index, df.mean_advantage, 'b-', label='mean of differential containment (MAG vs GTDB)')
plt.plot(df.index, df.max_cont, 'r.', label='best MAG containment across all metags', alpha=0.2)
plt.plot([0, max(df.index)], [0, 0], 'g--', label='(0 line: no advantage from MAGs)')
plt.legend(loc='lower left')
plt.xlabel('MAG rank (sorted by advantage)')
plt.ylabel('avg advantage of MAG containment vs best GTDB match')
plt.savefig('advantage.png')
```
